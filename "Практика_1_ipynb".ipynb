## **4. Визуализация распределения переменных**

### **Описание:**
Создайте визуализации для понимания распределения числовых и категориальных переменных.

### **Шаги:**


1. **Гистограмма для числовых переменных:**

```python
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
num_cols = [col for col in num_cols if col not in ['Weekdaysort', 'Monthsort']]
print(f"Числовые переменные для анализа: {list(num_cols)}")

if len(num_cols) == 0:
    print('В данных нет подходящих числовых переменных')
else:
    n_cols = len(num_cols)
    fig, axes = plt.subplots(1, n_cols, figsize=(6 * n_cols, 5))

    if n_cols == 1:
        axes = [axes]

    for ax, col in zip(axes, num_cols):
        sns.histplot(df[col], kde=True, ax=ax)
        ax.set_title(f'Распределение {col}')
        ax.set_xlabel(col)
        ax.set_ylabel('Частота')
    plt.tight_layout()
    plt.show()
```
